```python
import networkx as nx
from networkx.generators.random_graphs import erdos_renyi_graph


# Hello word graph design example with networkx

G0 = nx.Graph()

G0.add_edge(1, 2) # default edge data=1
G0.add_edge(2, 3)
G0.add_edge(1, 3)
G0.add_edge(1, 4)
G0.add_edge(4, 5)
G0.add_edge(4, 6)


nx.draw(G0)
```
