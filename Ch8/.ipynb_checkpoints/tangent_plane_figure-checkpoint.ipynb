```python
# written by Yi Rong from Boston University, yirong@bu.edu

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# customize figure size
plt.rcParams['figure.figsize'] = (10, 6)

# Make tangent plane
def f(x, y):
    return 1/np.exp(x ** 2 + y ** 2)

x = np.linspace(-1.5, 1.5, 30)
y = np.linspace(-1.5, 1.5, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the surface.
ax.plot_surface(X, Y, Z, 
                cmap=cm.summer,
                linewidth=0, antialiased=False)


# Make Curve Surface
def f(x, y):
    return 1.123 - 0.672 * x - 0.08 * y

x = np.linspace(-0, 0.8, 30)
y = np.linspace(-0.6, 1.15, 30)


X, Y = np.meshgrid(x, y)
Z = f(X, Y)
# Plot the surface.
ax.plot_surface(X, Y, Z, 
#                 cmap=cm.Blues,
                color='grey',
                linewidth=0, antialiased=False, 
                alpha=0.1)

#annotate
ax.scatter(0.3927, 0.047, 0.85519859, color='r', s=120) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

#camera position
ax.view_init(elev=20., azim=86)

# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)

plt.savefig('tangent_plane.png')
plt.show()
```
