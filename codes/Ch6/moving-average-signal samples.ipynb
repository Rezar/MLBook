```python
import numpy as np
import pandas as pd
from numpy import sqrt
#from matplotlib import pyplot

import pylab as plt


n = 1000
limit_low = 0
limit_high = 0.5
my_data = np.random.normal(0, 0.5, n) \
          + np.abs(np.random.normal(0, 2, n) \
                   * np.sin(np.linspace(0, 3*np.pi, n)) ) \
          + np.sin(np.linspace(0, 5*np.pi, n))**2 \
          + np.sin(np.linspace(1, 6*np.pi, n))**2

scaling = (limit_high - limit_low) / (max(my_data) - min(my_data))
my_data = my_data * scaling
my_data = my_data + (limit_low - min(my_data))

df = pd.DataFrame(my_data) 

rolling = df.rolling(window=50)
rolling_mean = (rolling.mean())


fig, ax1 = plt.subplots(figsize=(6,3))
ax1.plot(df,  color='blue')
ax2 = ax1.twiny()
ax2.plot(rolling_mean, color='red')

fig.tight_layout()
plt.show()

# https://scipy-cookbook.readthedocs.io/items/SignalSmooth.html 
```

```python
import matplotlib.pyplot as plt
import numpy as np
# if using a jupyter notebook
%matplotlib inline  


x = np.arange(-7,2.5*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)

         
#----- draw axis in the middle of the page -----
#spine placement data centered
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.yticks(np.arange(-1, 1, step=0.5))

#plt.plot(x,y,x,z)
plt.plot(x,y, color ='red')
plt.plot(x,z , color ='blue')
plt.savefig("High resoltion Sin Cosine.png",dpi=300)

plt.show()
```
