```python
#------ Plot sample PDF/PMF and CDF -------
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
  
# Plot between -100 and 100 with 0.1 steps.
gaus_data = np.arange(-100, 100, 0.1)
  
# Calculating mean and standard deviation
mean = statistics.mean(gaus_data)
sd = statistics.stdev(gaus_data)

#---- plot PDF -----
pdf = norm.pdf(gaus_data, mean, sd)
plt.plot(gaus_data, pdf, label="PDF")
plt.show()

#---- plot CDF -----
cdf = np.cumsum(pdf)

plt.plot( cdf, label="CDF")
plt.legend()
plt.show()
```

```python
#--------- Boltzmann distribution -----------
#source: https://notebook.community/tommyogden/quantum-python-lectures/11_Monte-Carlo-Maxwell-Boltzmann-Distributions 

import matplotlib.pyplot as plt
import numpy as np


def MB_speed(v,m,T):
    """ Maxwell-Boltzmann speed distribution for speeds """
    kB = 1.38e-23
    return (m/(2*np.pi*kB*T))**1.5 * 4*np.pi * v**2 * np.exp(-m*v**2/(2*kB*T))

fig = plt.figure()
ax = fig.add_subplot(111)

v = np.arange(0,800,1)
amu = 1.66e-27
mass = 85*amu

for T in [100,200,300]:
    fv = MB_speed(v,mass,T)
    ax.plot(v,fv,label='T='+str(T)+' K',lw=2)

ax.legend(loc=0)
ax.set_xlabel('$v$ (m/s)')
ax.set_ylabel('PDF, $f_v(v)$')
plt.savefig('Boltzmann.png', dpi=300)
plt.plot()
```

```python
# ------ Comparison between JS-divergence and KL-divergence ----
# source: https://qiita.com/Rowing0914/items/2f3fcb114ae671c7c394

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, entropy

def entropy_multi(p, q):
  return np.sum(p * np.log(p / q), axis=0)

def entropy_single(p):
  return np.sum(p * np.log(p), axis=0)

def KLD(pk, qk):
  # arraynise
  pk = np.asarray(pk)
  # normalise
  pk = 1.0*pk / np.sum(pk, axis=0)
  # check to decide if we apply single or multi entorpy
  if qk is None:
      return np.sum(entropy_single(pk), axis=0)
  else:
      # arraynise
      qk = np.asarray(qk)
      if len(qk) != len(pk):
          raise ValueError("qk and pk must have same length.")
      qk = 1.0*qk / np.sum(qk, axis=0)
      return np.sum(entropy_multi(pk, qk), axis=0)

# data prep
x = np.linspace(-10.0, 10.0, 1000)

# graph setting
plt.figure(figsize=(14,8))

# gradually shift the distribution
for i in np.arange(3):
    for j in np.arange(3):
        # index to shift
        index = i*3 + j
        # probabilistic distribution function
        p = norm.pdf(x, loc=-1, scale=1)
        q = norm.pdf(x, loc=index*0.8, scale=1.2)
        # mean of them
        m = (p+q)/2
        # scipy entropy
        kl = entropy(p, q)
        kl_mine = KLD(p,q)
        kl_mine2=KLD(q,p)
        # checker
        print("KL (p,q): ", "{0:.2f}".format(kl_mine), "KL (q,p) ", "{0:.2f}".format(kl_mine2))
        # prep for js divergence
        kl_pm = KLD(p, m)
        kl_qm = KLD(q, m)
        js = (kl_pm + kl_qm)/2
        plt.subplot(3,3,i*3+j+1)
        #plt.fill_between(x, m, facecolor="y", alpha=0.2)
        #plt.grid(False, color="white")
        
        plt.fill_between(x, p, facecolor="b", alpha=0.2)
        plt.fill_between(x, q, facecolor="r", alpha=0.2)
        plt.xlim(-5, 7)
        plt.ylim(0,0.45)
        plt.title("KL-Div:{:>.3f}".format(kl) + ",  JS-Div:{:>.3f}".format(js))
        plt.tick_params(labelbottom="off")
        plt.tick_params(labelleft="off")

plt.subplots_adjust(wspace=0.1, hspace=0.5)
plt.savefig('KL-JD-Divs.png', dpi=300)
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

#------- Simple equation to calculate KL-div -------
def kl_divergence(p, q):
   
    # normalise
    p = 1.0*p / np.sum(p, axis=0)
    return np.sum(np.where(p != 0, p * np.log(p / q), 0)) #--- Here we do additional check to not divide by zero. which is honestly incorrect 

#------- 

def js_divergence(p,q):
    js = (kl_divergence(p, q) + kl_divergence(q, p))/2
    return js


plt.figure(figsize=(18,5))

x = np.arange(-10, 10, 0.0001)
p = norm.pdf(x, -4, 1)
q = norm.pdf(x, 4, 1.1)
plt.fill_between(x, p, facecolor="b", alpha=0.2)
plt.fill_between(x, q, facecolor="r", alpha=0.2)

#plt.plot(x, p)
#plt.plot(x, q, c='red')

print("KL(P||Q):" , str(kl_divergence(p, q)) )
print("KL(Q||P):" , str(kl_divergence(q, p)) )
print("JS(Q||P) or JS(P||Q):", js_divergence(p,q))
```
