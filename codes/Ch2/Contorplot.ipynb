```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points with fewer data points
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# Simulate some elevation data for two mountains
Z = np.exp(-((X-2)**2 + (Y-2)**2)) + np.exp(-((X+2)**2 + (Y+2)**2))

# Create the contour plot
plt.figure(figsize=(8, 6))
cp = plt.contour(X, Y, Z, colors='white')  # contour lines
plt.clabel(cp, inline=True, fontsize=8)
plt.contourf(X, Y, Z, cmap='summer')  # filled contour
plt.colorbar()  # Adds a color bar to the side
plt.title('Contour Plot of Two Mountains')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Showing the grid
plt.grid(True)

plt.show()

```
