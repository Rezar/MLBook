```python
"""
Title: a complex shape for kNN classification
Author: Yi Rong from Boston University
Contact: yirong@bu.edu
"""

import matplotlib.pyplot as plt
from sklearn import neighbors
from mlxtend.plotting import plot_decision_regions

# customize figure size
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Generate a 2D Dataset and its labels
from sklearn.datasets import make_moons
x, y = make_moons(n_samples=100, noise=0.2, random_state=3) # increase noise if need more interleaving data
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors=1)
clf.fit(x, y)

# Plotting decision boundary
plot_decision_regions(x, y, clf=clf, legend=1, colors='red,blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('A Complex Shape for KNN with K=1')
plt.show()
```
