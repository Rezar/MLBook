```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float

def add_speckle_noise(image, mean=0, var=0.9):
    """
    Add speckle noise to image.
    The speckle noise is multiplicative, following the model:
    noisy = image + image * noise, where noise ~ N(mean, var)
    """
    row, col, ch = image.shape
    gauss = np.random.normal(mean, var ** 0.5, (row, col, ch))
    noisy = image + image * gauss
    return np.clip(noisy, 0, 1)

# Load and convert image to float
image = img_as_float(io.imread('/Users/reza/EVERYTHING/Work/ML Book/Figs/CH16/SARRadar.png'))

# Add speckle noise
noisy_image = add_speckle_noise(image)

# Display results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

ax1.imshow(image)
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(noisy_image)
ax2.set_title('Image with Speckle Noise')
ax2.axis('off')

plt.tight_layout()
plt.show()
```
