```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sinusoidal signal
t = np.linspace(0, 1, 500, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t)

# Crop the signal to a smaller amplitude
cropped_signal = np.clip(signal, -0.5, 0.5)

# Plot the original and cropped signals
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.ylim(-1.1, 1.1)

plt.subplot(2, 1, 2)
plt.plot(t, cropped_signal)
plt.title('Cropped Signal')
plt.ylim(-1.1, 1.1)

plt.tight_layout()
plt.show()
```
