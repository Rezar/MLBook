```python
import neurokit2 as nk  # Load the package

simulated_ecg = nk.ecg_simulate(duration=5, sampling_rate=50, heart_rate=80)
nk.signal_plot(simulated_ecg, sampling_rate=20)  # Visualize the signal

print(len(simulated_ecg))
```

```python
from matplotlib import pyplot as plt
import numpy as np 

plt.xlabel("Times") 
plt.ylabel("Heart Rate") 
plt.plot(simulated_ecg) 
#plt.rcParams["figure.dpi"] = 100
plt.show()
```

```python
import scipy.fftpack

# convert the Time signal into Frequency Signal with FFT

# Number of samplepoints
N = 250

# sample spacing
T = 1.0 / 250.0

yf = scipy.fftpack.fft(simulated_ecg)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```

```python
import pywt

pywt.families()

[phi, psi, x] = pywt.Wavelet('db2').wavefun(level=5)
plt.plot(x, psi)
plt.show()
```

```python
wavelet = pywt.ContinuousWavelet('gaus1')
psi, x = wavelet.wavefun(level=6)
plt.plot(x, psi)

plt.show()
```

```python
wavelet = pywt.ContinuousWavelet('mexh')
psi, x = wavelet.wavefun(level=12)
plt.plot(x, psi)

plt.show()

```
