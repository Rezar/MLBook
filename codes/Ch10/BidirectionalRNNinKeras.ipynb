```python
import numpy as np
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Dropout, Embedding, GRU, Bidirectional
from keras.datasets import imdb # IMDB movie reviews are embedded dataset in Keras
```

```python
n_unique_words = 10000 # cut texts after this number of words
maxlen = 200
batch_size = 64 
```

```python
(x_train, y_train),(x_test, y_test) = imdb.load_data(num_words=n_unique_words)
```

```python
 x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
 x_test = sequence.pad_sequences(x_test, maxlen=maxlen)
 y_train = np.array(y_train)
 y_test = np.array(y_test) 
```

```python
 model = Sequential()
 model.add(Embedding(n_unique_words, 128, input_length=maxlen))
 model.add(Bidirectional(GRU(64)))
 model.add(Dropout(0.5))
 model.add(Dense(1, activation='sigmoid'))
 model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
model.fit(x_train, y_train,batch_size=batch_size,epochs=3,validation_data=[x_test, y_test])

score = model.evaluate(x_test, y_test, verbose=0)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```
